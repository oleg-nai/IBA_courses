### Concat

```python
# Create tables for examples

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])


df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[4, 5, 6, 7])


df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},               
                    index=[8, 9, 10, 11])
```

```python
result = pd.concat([df1, df2, df3])
```

```python
result
```

```python
# Adding keys for simplifying indexing

result = pd.concat([df1, df2, df3], keys=['x', 'y', 'z'])
```

```python
result
```

```python
result.loc['y']
```

```python
# Creating new frame

df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
```

```python
df1
```

```python
df4
```

```python
# Check outer join

result = pd.concat([df1, df4], axis=1, join='outer', sort=False)
result
```

```python
# Check inner join

result = pd.concat([df1, df4], axis=1, join='inner', sort=False)
result
```

```python
# Check how ignore_index works

result = pd.concat([df1, df4], ignore_index=True, sort=False)
result
```

### Merge

```python
# Create frames

left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})


right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
# Merge on 2 keys with default parameters

result = pd.merge(left, right, on=['key1', 'key2'])
result
```

```python
# LEFT merge

result = pd.merge(left, right, how='left', on=['key1', 'key2'])
print('Left')
result
```

```python
# RIGHT merge

result = pd.merge(left, right, how='right', on=['key1', 'key2'])
print('Right')
result
```

### Join

```python
# Create examples of frames

left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                    index=['K0', 'K1', 'K2'])


right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                     index=['K0', 'K2', 'K3'])

result = left.join(right)
result
```
